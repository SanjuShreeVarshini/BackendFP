```python
# STEP 1: Upload CSV
from google.colab import files
files.upload()

# STEP 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# STEP 3: Load data
df = pd.read_csv("student_scores.csv")

# STEP 4: Show data
display(df)

# ---------------- BASIC CHARTS ----------------

# Bar Chart  Math Marks
plt.figure()
plt.bar(df["Name"], df["Math"])
plt.xticks(rotation=45)
plt.xlabel("Student")
plt.ylabel("Math Marks")
plt.title("Math Scores of Students")
plt.show()

# Line Chart  Hours Studied vs Math
plt.figure()
plt.plot(df["Hours_Studied"], df["Math"], marker="o")
plt.xlabel("Hours Studied")
plt.ylabel("Math Marks")
plt.title("Hours Studied vs Math Marks")
plt.show()

# Scatter Plot  Science vs English
plt.figure()
plt.scatter(df["Science"], df["English"])
plt.xlabel("Science Marks")
plt.ylabel("English Marks")
plt.title("Science vs English Marks")
plt.show()

# Histogram  English Marks
plt.figure()
plt.hist(df["English"], bins=5)
plt.xlabel("English Marks")
plt.ylabel("No. of Students")
plt.title("Distribution of English Marks")
plt.show()

# ---------------- ADVANCED CHARTS ----------------

# Box Plot  Subject Score Distribution
plt.figure()
sns.boxplot(data=df[["Math", "Science", "English"]])
plt.title("Box Plot of Subject Scores")
plt.show()

# Violin Plot  Subject Score Density
plt.figure()
sns.violinplot(data=df[["Math", "Science", "English"]])
plt.title("Violin Plot of Subject Scores")
plt.show()

# KDE Plot  Math Score Distribution
plt.figure()
sns.kdeplot(df["Math"], fill=True)
plt.title("KDE Plot of Math Scores")
plt.show()

# Count Plot  Study Hours Frequency
plt.figure()
sns.countplot(x="Hours_Studied", data=df)
plt.title("Frequency of Study Hours")
plt.show()

# ---------------- COMPARISON CHARTS ----------------

# Average marks per subject
avg_scores = df[["Math", "Science", "English"]].mean()

plt.figure()
avg_scores.plot(kind="bar")
plt.ylabel("Average Marks")
plt.title("Average Marks per Subject")
plt.show()

# ---------------- HEATMAP (FIXED) ----------------

# Correlation Heatmap (NUMERIC ONLY)
plt.figure()
numeric_df = df.select_dtypes(include="number")
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
