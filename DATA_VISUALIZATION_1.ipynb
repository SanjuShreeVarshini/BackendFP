```python
# STEP 1: Upload CSV
from google.colab import files
files.upload()

# STEP 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# STEP 3: Load data
df = pd.read_csv("student_scores.csv")
display(df)

# ---------------- BAR CHART ----------------
# Math Marks per Student
plt.figure()
plt.bar(df["Name"], df["Math"], color="skyblue", edgecolor="black")
plt.xticks(rotation=45)
plt.xlabel("Student Name")
plt.ylabel("Math Marks")
plt.title("Math Scores of Students")
plt.show()

# Correct Line Chart  Hours Studied vs Math

df_sorted = df.sort_values(by="Hours_Studied")

plt.figure()
plt.plot(
    df_sorted["Hours_Studied"],
    df_sorted["Math"],
    color="red",
    marker="o",
    linewidth=2
)

plt.xlabel("Hours Studied")
plt.ylabel("Math Marks")
plt.title("Hours Studied vs Math Marks")
plt.grid(True)
plt.show()


# ---------------- SCATTER CHART ----------------
# Science vs English
plt.figure()
plt.scatter(df["Science"], df["English"],
            color="green", s=80)
plt.xlabel("Science Marks")
plt.ylabel("English Marks")
plt.title("Science vs English Marks")
plt.show()

# ---------------- HISTOGRAM ----------------
# English Marks Distribution
plt.figure()
plt.hist(df["English"], bins=5,
         color="orange", edgecolor="black")
plt.xlabel("English Marks")
plt.ylabel("Number of Students")
plt.title("English Marks Distribution")
plt.show()

# ---------------- BOX PLOT (SMALLER WIDTH) ----------------
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df[["Math", "Science", "English"]],
    palette="Set2",
    width=0.4   #  makes boxes thinner
)
plt.title("Box Plot of Subject Scores")
plt.show()


# ---------------- VIOLIN PLOT ----------------
plt.figure()
sns.violinplot(
    data=df[["Math", "Science", "English"]],
    palette="Pastel1"
)
plt.title("Violin Plot of Subject Scores")
plt.show()

# ---------------- KDE PLOT ----------------
plt.figure()
sns.kdeplot(df["Math"], fill=True, color="purple")
plt.title("KDE Plot of Math Marks")
plt.show()

# ---------------- COUNT PLOT ----------------
plt.figure()
sns.countplot(x="Hours_Studied", data=df, palette="viridis")
plt.title("Frequency of Study Hours")
plt.show()

# ---------------- AVERAGE MARKS ----------------
avg_marks = df[["Math", "Science", "English"]].mean()

plt.figure()
avg_marks.plot(kind="bar", color=["blue", "green", "red"])
plt.ylabel("Average Marks")
plt.title("Average Marks per Subject")
plt.show()
# ---------------- 3D SCATTER PLOT ----------------
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    df["Math"],
    df["Science"],
    df["English"],
    color="purple",
    s=60
)

ax.set_xlabel("Math Marks")
ax.set_ylabel("Science Marks")
ax.set_zlabel("English Marks")
ax.set_title("3D Scatter Plot of Student Scores")

plt.show()
# ---------------- 3D LINE PLOT ----------------
df_sorted = df.sort_values(by="Math")

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot(
    df_sorted["Math"],
    df_sorted["Science"],
    df_sorted["English"],
    color="red",
    marker="o"
)

ax.set_xlabel("Math Marks")
ax.set_ylabel("Science Marks")
ax.set_zlabel("English Marks")
ax.set_title("3D Line Plot of Student Scores")

plt.show()
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create a grid
x = df["Math"]
y = df["Science"]
z = df["English"]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Surface requires meshgrid
X, Y = np.meshgrid(np.linspace(x.min(), x.max(), 20),
                   np.linspace(y.min(), y.max(), 20))

# Example: simple interpolation for Z
Z = np.outer(np.linspace(z.min(), z.max(), 20), np.ones(20))

ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

ax.set_xlabel("Math Marks")
ax.set_ylabel("Science Marks")
ax.set_zlabel("English Marks")
ax.set_title("3D Surface Plot of Scores")

plt.show()
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

x = np.arange(len(df))  # index for students
y = df["Math"]
z = np.zeros(len(df))   # bars start from 0

dx = dy = 0.5           # width of bars
dz = df["Science"]      # height in z-axis (could be English or Science)

ax.bar3d(x, y, z, dx, dy, dz, color='orange', alpha=0.8)

ax.set_xlabel("Student Index")
ax.set_ylabel("Math Marks")
ax.set_zlabel("Science Marks")
ax.set_title("3D Bar Plot of Student Scores")

plt.show()
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.stem(df["Math"], df["Science"], df["English"], linefmt='grey', markerfmt='ro', basefmt=" ")

ax.set_xlabel("Math Marks")
ax.set_ylabel("Science Marks")
ax.set_zlabel("English Marks")
ax.set_title("3D Stem Plot of Scores")

plt.show()

# ---------------- HEATMAP ----------------
plt.figure()
sns.heatmap(
    df.select_dtypes(include="number").corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```
