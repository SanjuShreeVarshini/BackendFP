```python
# ================================
# INSTALL LIBRARIES
# ================================
!pip -q install ultralytics opencv-python matplotlib

# ================================
# IMPORTS
# ================================
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from ultralytics import YOLO

# ================================
# UPLOAD IMAGE
# ================================
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ================================
# PREPROCESSING
# ================================
img_resized = cv2.resize(img, (640, 640))
gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
normalized = gray / 255.0

# ================================
# IMAGE ENHANCEMENT
# ================================
equalized = cv2.equalizeHist(gray)

# ================================
# DENOISING
# ================================
gaussian = cv2.GaussianBlur(equalized, (5,5), 0)
median = cv2.medianBlur(equalized, 5)

# ================================
# EDGE ENHANCEMENT
# ================================
edges = cv2.Canny(gaussian, 100, 200)

# ================================
# OBJECT DETECTION (YOLOv8)
# ================================
model = YOLO("yolov8n.pt")
results = model(img_rgb)
detected_img = results[0].plot()

# ================================
# DISPLAY RESULTS
# ================================
titles = [
    "Original", "Grayscale", "Enhanced",
    "Gaussian Denoised", "Median Denoised",
    "Edges", "Object Detection"
]

images = [
    img_rgb, gray, equalized,
    gaussian, median,
    edges, detected_img
]

plt.figure(figsize=(15,10))
for i in range(len(images)):
    plt.subplot(2,4,i+1)
    if len(images[i].shape) == 2:
        plt.imshow(images[i], cmap='gray')
    else:
        plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.show()

# ================================
# SAVE OUTPUT
# ================================
cv2.imwrite("detected_output.jpg",
            cv2.cvtColor(detected_img, cv2.COLOR_RGB2BGR))

files.download("detected_output.jpg")
```
